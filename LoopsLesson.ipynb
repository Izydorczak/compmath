#### for loops are useful for doing something to or with each item in a list or other iterable. 

```
dogs = ['Rover', 'Max', 'Holly']        
for dog in dogs:
        dog = dog + '\'s'
        print(dog)              
```

#### Here is an example of a common routine. We start with an empty list and fill it as we loop through another list. Remember to guess the outcome before trying.

```
ages_later = []         
years = 10
ages = [12, 15, 17]
for age in ages:
        age +=years
        ages_later.append(age)
print(ages_later)
```
